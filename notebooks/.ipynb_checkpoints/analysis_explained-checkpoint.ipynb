## О чём вообще исследование (в двух абзацах)

**БТА (ботулинотерапия)** — это лечебная процедура: в стенку мочевого пузыря
через цистоскоп вводят ботулинический токсин типа А (тот же «ботокс», что в
косметологии, но здесь — лечебно). Так лечат *гиперактивный* или
*нейрогенный* мочевой пузырь, когда пузырь сокращается непроизвольно и
человек не удерживает мочу. Это инвазивная манипуляция через мочевые пути,
поэтому есть риск занести/спровоцировать **инфекцию мочевыводящих путей
(ИМВП)**.

Чтобы снизить этот риск, перед процедурой назначают **антибиотикопрофилактику**
(профилактический антибиотик) и заранее проверяют мочу на инфекцию. Наше
исследование оценивает: как часто после БТА развивается ИМВП, насколько
согласованы два способа проверки мочи, и какие организационные проблемы есть
в скрининге. Выборка маленькая (десятки процедур), поэтому особое внимание —
честной оценке неопределённости.

## 📖 Словарь терминов (можно возвращаться сюда по ходу)

**Клинические / лабораторные**
- **БТА** — ботулинотерапия мочевого пузыря (внутридетрузорное введение
  ботулотоксина А). «Детрузор» — мышца стенки пузыря.
- **ИМВП** — инфекция мочевыводящих путей (по-английски UTI). Здесь —
  *симптомная* ИМВП в течение 5 суток после процедуры (главный исход).
- **ОАМ** — общий анализ мочи. Быстрый, дешёвый тест (лейкоциты, нитриты,
  бактерии «на глаз» под микроскопом). Скрининговый, но грубый.
- **Посев мочи / МИ** (микробиологическое исследование) — мочу высевают на
  питательную среду, бактерии вырастают за 1–3 суток. Точно показывает, какой
  это микроб и сколько его. Медленнее и дороже ОАМ, но чувствительнее.
- **Бактериурия** — наличие бактерий в моче. **Значимая бактериурия** —
  когда их выше порога (классически ≥10⁵ КОЕ/мл), т.е. это рост, а не
  случайное загрязнение пробы.
- **КОЕ/мл** (колониеобразующие единицы на мл) — мера «сколько живых бактерий».
  Часто берут **log10 КОЕ**: 3 → тысяча, 5 → сто тысяч, 6 → миллион/мл.
- **Дренирование** — как у пациента отводится моча (важный фактор риска):
  - *intermittent* — периодическая самокатетеризация (катетер вводят и
    убирают несколько раз в день);
  - *indwelling* — постоянный уретральный катетер (стоит всё время);
  - *cystostomy* — цистостома, катетер через прокол над лобком напрямую в
    пузырь;
  - *pads* — без катетера, прокладки при недержании.
- **Резистентность** — устойчивость выявленного микроба к антибиотикам.
- **Режимы профилактики**: *single* — однократная доза; *combined* —
  комбинация препаратов; *prolonged* — пролонгированный (продлённый) курс.
- **Диссоциация ОАМ/посев** — расхождение: ОАМ говорит «чисто», а посев
  находит рост. Ключевая находка исследования.

**Статистические**
- **Медиана** — «серединное» значение (половина наблюдений меньше, половина
  больше). Устойчивее среднего к выбросам.
- **IQR** (межквартильный размах) — диапазон от 25-го до 75-го процентиля,
  т.е. где лежат «средние» 50% наблюдений.
- **Цензурирование** — пациент выбыл из наблюдения раньше 5 суток, поэтому мы
  *не знаем*, была у него ИМВП или нет. Таких исключаем из первичной оценки,
  чтобы не врать.
- **Первичная конечная точка** — заранее объявленный главный исход
  (здесь: ИМВП за 5 суток). Всё остальное — вторичное/поисковое.
- **Доверительный интервал (CI95)** — диапазон, в котором правдоподобно лежит
  истинная доля. **Wilson** и **Clopper-Pearson** — два способа его построить
  (CP «точный» и осторожный — даёт шире).
- **Байес / Beta-Binomial / HDI** — байесовская оценка доли. Даёт
  *распределение* правдоподобия доли, а не одну точку. **HDI95** — интервал
  наибольшей плотности (самые правдоподобные 95%). Работает осмысленно даже
  при 1–2 событиях, где частотные методы «ломаются».
- **McNemar (тест Мак-Нимара)** — парный тест для таблицы 2×2: сравнивает два
  теста (ОАМ и посев) **на одних и тех же** процедурах. Смотрит на *расхождения*.
- **Wilcoxon (знаково-ранговый)** — парный непараметрический тест: отличаются
  ли сроки ОАМ и посева, без предположения о нормальности.
- **Точный тест Фишера** — для маленьких таблиц 2×2: связан ли фактор с исходом.
- **Odds Ratio (OR, отношение шансов)** — во сколько раз меняются шансы исхода
  при наличии фактора. OR=1 — нет связи; <1 — фактор «защищает»; >1 —
  повышает риск.
- **p-value** — вероятность увидеть такие (или более крайние) данные, если
  эффекта на самом деле нет. Маленькое p (<0.05) — «вряд ли случайность».
- **Статистическая мощность** — способность исследования заметить эффект, если
  он есть. При малом числе событий мощность низкая → «незначимо» НЕ значит
  «эффекта нет».

## 1. Запуск пайплайна — «единый источник чисел»

Одна функция `pipeline.run(...)` прогоняет весь конвейер по порядку:

1. **ingest** — читает сырые Excel-файлы (`data/raw/*.xlsx`).
2. **features** — извлекает признаки из текста (возраст, пол, даты, организмы,
   результаты ОАМ/посева, сроки до БТА и т.д.).
3. **dedup** — убирает дубли. *Единица анализа* — **процедура**: один пациент
   мог пройти БТА несколько раз, и каждая процедура считается отдельно, но
   повторно занесённые одинаковые строки удаляются.
4. **validate** — проверяет, что таблица соответствует «контракту» (нужные
   колонки, типы, диапазоны). Артефакты данных ловятся здесь, а не глазами.
5. **stats** — считает всю статистику и складывает в один объект `res`.

`res.provenance` — «происхождение чисел»: сколько строк было сырых, сколько
осталось после дедупликации, сколько исключено по возрасту. Это прозрачность:
видно, откуда взялось итоговое N.

```python
res = pipeline.run(ROOT / "data" / "raw", ROOT / "config" / "protocol.yaml")
df = res.analytical            # финальная аналитическая таблица: одна строка = одна процедура
print("Происхождение чисел:", res.provenance)
df.head()
```

**Как читать `df`:** каждая строка — одна процедура БТА. Полезные колонки:
`age`/`sex` — возраст/пол; `drainage` — тип дренирования; `proph_regimen`/
`proph_drug` — режим и препарат профилактики; `oam_positive`/`mi_positive` —
нашёл ли рост ОАМ / посев; `max_log_cfu` — бактериальная нагрузка по посеву;
`oam_to_bta_days`/`mi_to_bta_days` — за сколько дней до процедуры сделали тест;
`uti_5d` — была ли ИМВП за 5 суток; `censored` — выбыл ли пациент до 5 суток.

дате БТА — иначе результат «протухает» (инфекция могла появиться позже).

Мы сравниваем, **за сколько дней до БТА** в среднем выполняют ОАМ и посев.
`oam_to_bta_days` и `mi_to_bta_days` — эти интервалы. Сравниваем медианы и

```python
t = res.timing
print(f"ОАМ медиана {t['oam_median']:.0f} сут, посев медиана {t['mi_median']:.0f} сут")
if "wilcoxon_p" in t:
    print(f"Парный Wilcoxon p = {t['wilcoxon_p']:.4g}")
```

**Boxplot («ящик с усами»):** коробка — это IQR (средние 50%), линия внутри —
медиана, усы — разброс, точки за усами — выбросы. Видно: посев делают **сильно

## 4. Диссоциация ОАМ и посева — главная клиническая находка

Вопрос: **согласованы ли** ОАМ и посев? Берём процедуры, где сделали **оба**
теста, и строим таблицу 2×2: для каждой процедуры — что показал ОАМ (+/−) и
что показал посев (+/−).

Нас интересуют **расхождения** (диссонанс):
- **ОАМ−/посев+** — ОАМ сказал «чисто», а посев нашёл рост → **ОАМ пропустил
  инфекцию**. Это клинически опасный тип ошибки.
- **ОАМ+/посев−** — обратное (бывает редко).

**Тест Мак-Нимара** проверяет именно асимметрию расхождений: систематически
ли один тест «пропускает» относительно другого. Маленькое p → расхождение не
случайно.

```python
mc = res.mcnemar
print(f"Пар с обоими тестами: {mc['n_pairs']}")
print(f"ОАМ−/посев+ : {mc['oam_neg_mi_pos']}   (ОАМ пропустил рост)")
print(f"ОАМ+/посев− : {mc['oam_pos_mi_neg']}")
print(f"McNemar exact p = {mc['p_value']:.4g}")
print(mc["discordance_note"])
```

Таблица сопряжённости (строки — что сказал ОАМ, столбцы — что сказал посев).
Берём её из поля `contingency` (корректно ориентированная сетка из ядра).

```python
ct = pd.DataFrame(mc["contingency"],
                  index=["ОАМ+", "ОАМ−"], columns=["посев+", "посев−"])
ct
```

Та же таблица как **тепловая карта** — нагляднее. Тёмная клетка «ОАМ−/посев+»
и есть инфекции, которые **ОАМ пропустил**, а посев поймал.

```python
grid = np.array(mc["contingency"])
fig, ax = plt.subplots(figsize=(5.5, 4.5))
im = ax.imshow(grid, cmap="Blues")
ax.set_xticks([0, 1], ["посев +", "посев −"])
ax.set_yticks([0, 1], ["ОАМ +", "ОАМ −"])
for i in range(2):
    for j in range(2):
        ax.text(j, i, grid[i, j], ha="center", va="center", fontsize=18,
                color="white" if grid[i, j] > grid.max() / 2 else "black")
ax.set_title(f"Диссоциация ОАМ vs посев (n={mc['n_pairs']} пар)\n"
             f"ОАМ пропустил рост: {mc['oam_neg_mi_pos']} | "
             f"McNemar p = {mc['p_value']:.2g}")
plt.tight_layout(); plt.show()
```

### Почему так? Масштаб роста, который ОАМ не видит

ОАМ — грубый тест; он пропускает рост, особенно при «нетипичной» флоре. Здесь
смотрим **бактериальную нагрузку по посеву** (`max_log_cfu`, в log10 КОЕ/мл),
разбитую по тому, что сказал ОАМ. Если при отрицательном ОАМ посев всё равно
показывает высокую нагрузку — это прямое объяснение диссоциации.

```python
fig, ax = plt.subplots(figsize=(7, 4))
sub = df.dropna(subset=["max_log_cfu"])
bins = np.arange(0, sub["max_log_cfu"].max() + 2) - 0.5
for val, lab, color in [(False, "ОАМ отрицателен", "#C44E52"),
                        (True, "ОАМ положителен", "#4C72B0")]:
    d = sub.loc[sub["oam_positive"] == val, "max_log_cfu"]
    ax.hist(d, bins=bins, alpha=0.6, label=f"{lab} (n={len(d)})", color=color)
ax.set_xlabel("log10 КОЕ/мл (макс. по посеву)")
ax.set_ylabel("Процедур")
ax.set_title("Бактериальная нагрузка по посеву vs результат ОАМ")
ax.legend()
plt.tight_layout(); plt.show()
```

## 5. Первичная конечная точка — частота ИМВП за 5 суток

Главный исход: **доля процедур, после которых развилась симптомная ИМВП в
течение 5 суток**. Из расчёта исключаем **цензурированных** (выбыли раньше 5
суток — про них исход неизвестен), иначе занизим/исказим оценку.

Сложность: событий очень мало (единицы). При малом числе обычная формула
«события ÷ всего» даёт обманчиво узкую картину. Поэтому показываем долю **тремя
способами**:
- **Wilson** — хороший общий CI для доли;
- **Clopper-Pearson** — «точный», осторожный (интервал шире — честнее при
  малых N);
- **Bayes Beta(1,1)** — байес с равномерным априором; даёт распределение и
  **HDI**, осмысленно даже при 1–2 событиях.

И в конце — **предупреждение о мощности**: при горстке событий незначимость
ничего не доказывает.

```python
p = res.primary
for label, key in [("Wilson", "uti_rate_wilson"), ("Clopper-Pearson", "uti_rate_exact")]:
    r = p[key]
    print(f"{label:16s}: {r['point']:.1%}  CI95 [{r['ci_low']:.1%}; {r['ci_high']:.1%}]")
b = p["uti_rate_bayes"]
print(f"{'Bayes Beta(1,1)':16s}: {b['posterior_mean']:.1%}  HDI95 [{b['hdi_low']:.1%}; {b['hdi_high']:.1%}]")
print()
print(stats.power_warning(p["uti_events"], p["n_evaluable"]))
```

**Апостериорное распределение (байес).** Кривая — насколько правдоподобно
каждое значение истинной частоты ИМВП. Пунктир — границы HDI95. Широкая кривая
= большая неопределённость (потому что событий мало).

```python
from scipy import stats as sps
a_post = 1 + p["uti_events"]
b_post = 1 + (p["n_evaluable"] - p["uti_events"])
xs = np.linspace(0, 0.3, 400)
plt.figure(figsize=(7, 4))
plt.plot(xs, sps.beta(a_post, b_post).pdf(xs))
plt.axvline(b["hdi_low"], ls="--", c="gray")
plt.axvline(b["hdi_high"], ls="--", c="gray")
plt.title("Апостериорное распределение частоты ИМВП (Beta-Binomial)")
plt.xlabel("Частота ИМВП за 5 суток"); plt.ylabel("Плотность")
plt.tight_layout(); plt.show()
```

**Forest-plot трёх методов** — три оценки бок о бок (точка = оценка, линия =
интервал). Видно, что все три согласуются по «точке», но Clopper-Pearson самый
широкий, а байес слегка сдвинут — это нормально при малом N.

```python
ests = [
    ("Wilson", p["uti_rate_wilson"]["point"],
     p["uti_rate_wilson"]["ci_low"], p["uti_rate_wilson"]["ci_high"]),
    ("Clopper-Pearson", p["uti_rate_exact"]["point"],
     p["uti_rate_exact"]["ci_low"], p["uti_rate_exact"]["ci_high"]),
    ("Bayes Beta(1,1)", p["uti_rate_bayes"]["posterior_mean"],
     p["uti_rate_bayes"]["hdi_low"], p["uti_rate_bayes"]["hdi_high"]),
]
fig, ax = plt.subplots(figsize=(7.5, 3))
for y, (lab, pt, lo, hi) in enumerate(ests):
    ax.plot([lo * 100, hi * 100], [y, y], color="#4C72B0", lw=2)
    ax.plot(pt * 100, y, "o", color="#4C72B0")
    ax.text(hi * 100 + 0.6, y, f"{pt:.1%} [{lo:.1%}; {hi:.1%}]",
            va="center", fontsize=9)
ax.set_yticks(range(len(ests)), [e[0] for e in ests])
ax.set_xlabel("Частота ИМВП за 5 суток, %")
ax.set_xlim(0, max(e[3] for e in ests) * 100 + 8)
ax.set_title(f"Оценки доли ИМВП ({p['uti_events']}/{p['n_evaluable']}) — три метода")
ax.invert_yaxis()
plt.tight_layout(); plt.show()
```

## 6. Факторы риска и исход (тест Фишера) — строго как гипотезы

Проверяем, связан ли каждый бинарный фактор с ИМВП:
- `any_prophylaxis` — была ли вообще профилактика;
- `resistant_noted` — отмечена ли резистентность микроба;
- `significant_bacteriuria` — была ли значимая бактериурия.

Для каждого — таблица 2×2 и **точный тест Фишера** (он корректен на малых
числах). Считаем **OR (отношение шансов)** и p.

⚠️ Важно: событий мало, поэтому **все** связи статистически незначимы, а
некоторые OR вырождаются в 0 (в одной из клеток таблицы просто нет событий).
Это не «доказательство отсутствия эффекта» — это **нехватка данных**. Трактуем
только как направления-гипотезы.

```python
fisher_df = pd.DataFrame([
    {"factor": k, "OR": v["odds_ratio"], "p": v["p_value"], "n": v["n"]}
    for k, v in res.fisher.items()
])
fisher_df
```

**Forest-plot OR на логарифмической шкале.** Пунктир на OR=1 — «нет эффекта».
Где OR обнуляется (нет событий в группе), честно ставим открытый маркер у края
с подписью «OR→0». Картинка специально показывает: оценки крайне неустойчивы.

```python
fac = pd.DataFrame([
    {"factor": k, "OR": v["odds_ratio"], "p": v["p_value"]}
    for k, v in res.fisher.items()
])
labels = {"any_prophylaxis": "Любая профилактика",
          "resistant_noted": "Отмечена резистентность",
          "significant_bacteriuria": "Значимая бактериурия"}
floor = 0.01
fig, ax = plt.subplots(figsize=(7.5, 3))
for y, row in enumerate(fac.itertuples()):
    orr = row.OR if (row.OR and row.OR > 0) else None
    if orr is None:
        ax.plot(floor, y, "o", mfc="white", mec="#C44E52")
        ax.text(floor * 1.4, y, f"OR→0 (нет событий), p={row.p:.2g}",
                va="center", fontsize=9)
    else:
        ax.plot(orr, y, "o", color="#4C72B0")
        ax.text(orr * 1.4, y, f"OR={orr:.2g}, p={row.p:.2g}",
                va="center", fontsize=9)
ax.axvline(1, ls="--", color="gray")
ax.set_xscale("log")
ax.set_xlim(floor * 0.5, 30)
ax.set_yticks(range(len(fac)), [labels.get(f, f) for f in fac["factor"]])
ax.set_xlabel("Odds Ratio (лог-шкала) — фактор vs ИМВП")
ax.set_title("Факторы и исход (Fisher) — все ассоциации незначимы")
ax.invert_yaxis()
plt.tight_layout(); plt.show()
```

## 7. Динамика микрофлоры у повторных пациентов

Часть пациентов проходила БТА несколько раз. Сравниваем, какие микробы росли
в моче от визита к визиту, и классифицируем **тип изменения**:
- **сохранение (persistence)** — тот же набор микробов;
- **приобретение (acquisition)** — к старым добавились новые;
- **частичная потеря (partial_loss)** — часть микробов ушла;
- **полная смена (full_switch)** — флора полностью сменилась;
- **появление (appearance)** — раньше было пусто, теперь есть рост;
- **элиминация (elimination)** — раньше рос, теперь чисто.

Это поисковое наблюдение: меняется ли колонизатор и как часто инфекция
«закрепляется».

```python
res.repeat
```

**Столбчатая диаграмма** — сколько переходов каждого типа. Преобладание
«сохранения»/«приобретения» говорило бы о стойкой колонизации.

```python
ru = {"persistence": "сохранение", "acquisition": "приобретение",
      "partial_loss": "частичная потеря", "full_switch": "полная смена",
      "appearance": "появление", "elimination": "элиминация"}
order = ["persistence", "acquisition", "partial_loss",
         "full_switch", "appearance", "elimination"]
cnt = res.repeat["change"].value_counts().reindex(order).dropna().astype(int)
fig, ax = plt.subplots(figsize=(7.5, 4))
ax.barh([ru[i] for i in cnt.index], cnt.values, color="#55A868")
for y, v in enumerate(cnt.values):
    ax.text(v + 0.05, y, int(v), va="center")
ax.set_xlabel("Число повторных визитов")
ax.set_title(f"Динамика микрофлоры у повторных пациентов (переходов: {len(res.repeat)})")
ax.invert_yaxis()
plt.tight_layout(); plt.show()
```

## 8. Главные выводы — простыми словами

1. **ОАМ ненадёжен как единственный скрининг.** Он систематически пропускает
   рост, который ловит посев (диссоциация ОАМ−/посев+, McNemar p очень мал).
   Иногда при «чистом» ОАМ нагрузка по посеву высокая.
2. **Посев делают слишком рано** относительно даты БТА → к моменту процедуры
   результат может устареть. Это организационная, исправимая проблема.
3. **Частота ИМВП за 5 суток низкая**, но событий мало — доверительные
   интервалы широкие, и точную цифру утверждать нельзя.
4. **Факторы риска не подтверждены статистически** — не потому что их нет, а
   потому что данных мало (низкая мощность). Это гипотезы для большего
   исследования.

`src/bta_study`, поэтому ноутбук, отчёт и слайды не разойдутся в цифрах.
