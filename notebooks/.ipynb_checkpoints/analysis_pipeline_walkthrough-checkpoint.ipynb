# Как работает пайплайн БТА — пошагово, от сырых данных к финальным

Этот ноутбук не повторяет анализ, а **показывает внутреннюю кухню**: как сырые
Excel-таблицы превращаются в чистую аналитическую таблицу и числа. Мы пройдём
конвейер «руками», вызывая каждый слой по отдельности и разглядывая, что
изменилось на каждом шаге.

```
   data/raw/*.xlsx
        │
        ▼
 ┌──────────────┐   1. INGEST     читаем Excel, переименовываем колонки,
 │   ingest     │                 склеиваем файлы в одну «длинную» таблицу
 └──────┬───────┘
        ▼
 ┌──────────────┐   2. FEATURES   парсим текст (даты, возраст, микробы…),
 │   features   │                 считаем интервалы и конечные точки
 └──────┬───────┘
        ▼
 ┌──────────────┐   3. DEDUP      убираем дубли процедур по ключу
 │    dedup     │                 (фамилия + дата БТА)
 └──────┬───────┘
        ▼
 ┌──────────────┐   4. ФИЛЬТР     критерий включения (возраст ≥ 18)
 │  inclusion   │
 └──────┬───────┘
        ▼
 ┌──────────────┐   5. VALIDATE   контракт данных (Pandera): падаем, если
 │   schema     │                 что-то не так — ДО статистики
 └──────┬───────┘
        ▼
 ┌──────────────┐   6. STATS      доли, McNemar, Wilcoxon, Fisher, динамика
 │    stats     │                 → единый объект Results (res)
 └──────────────┘
```

Принцип архитектуры: каждый слой — **чистая функция** (один вход → один
предсказуемый выход). Поэтому пайплайн воспроизводим, тестируем и его легко
разобрать по косточкам, что мы сейчас и сделаем.

## 0. Импортируем слои по отдельности

В обычном анализе мы зовём `pipeline.run(...)` — одну функцию, которая делает
всё. Здесь же импортируем **каждый модуль отдельно**, чтобы вызывать этапы
вручную и смотреть промежуточные результаты.

````python
from pathlib import Path
import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if (ROOT / "src").exists():
    sys.path.insert(0, str(ROOT))
elif (ROOT.parent / "src").exists():
    ROOT = ROOT.parent
    sys.path.insert(0, str(ROOT))

# слои пайплайна — каждый по отдельности
from src.bta_study import ingest, features, dedup, schema, stats, pipeline
from src.bta_study import parsers as P

pd.set_option("display.max_columns", 50)
RAW_DIR = ROOT / "data" / "raw"
print("Сырые файлы:", [p.name for p in sorted(RAW_DIR.glob("raw_*.xlsx"))])
````

## 1. INGEST — чтение сырых данных

Исходные данные — это «живые» Excel-таблицы из больницы, по одному файлу на
источник. Колонки названы по-русски и неудобно (`№ анализа`, `ИМВП в теч. 5
дней...`). Слой `ingest`:

1. читает все `raw_*.xlsx`,
2. **переименовывает** колонки в короткие технические имена (словарь
   `ingest.RENAME`),
3. **склеивает** файлы в одну таблицу, добавляя `source_file` (откуда строка).

Сначала посмотрим, как выглядит сырой файл **до** обработки.

````python
sample_path = sorted(RAW_DIR.glob("raw_*.xlsx"))[0]
raw0 = pd.read_excel(sample_path, dtype={"Дата БТА": object})
print(f"Файл {sample_path.name}: строк {len(raw0)}, колонок {raw0.shape[1]}")
print("Исходные (русские) названия колонок:")
list(raw0.columns)
````

Словарь переименования — «русское название → техническое имя»:

````python
ingest.RENAME
````

Теперь сам `ingest.load_raw` — он объединяет оба файла и применяет переименование:

````python
raw = ingest.load_raw(RAW_DIR)
print(f"Объединено в одну таблицу: {len(raw)} строк, {raw.shape[1]} колонок")
print("Технические имена колонок:", list(raw.columns))
raw.head(3)
````

## 2. FEATURES — парсинг текста в структурированные признаки

Это самый «умный» слой. В сырых данных много информации спрятано **внутри
текста**. Например, в одной ячейке ФИО лежит и фамилия, и дата рождения, и тип
дренирования:

> `АВДЕЕВ Е.П., 03.01.1974 г.р., ботулинотерапия ... цистостома`

Слой `features` пропускает каждую строку через набор парсеров (`parsers.py`) и
получает аккуратные колонки. Разберём это **на одной строке**, чтобы увидеть
магию вживую.

````python
# берём одну показательную сырую строку
mask = raw["name_cell"].astype(str).str.startswith("АВДЕЕВ")
r = raw[mask].iloc[0] if mask.any() else raw.iloc[0]

print("СЫРОЙ ТЕКСТ (что пришло из Excel):")
print("  ФИО-ячейка        :", repr(str(r["name_cell"])[:80]))
print("  Дата БТА (сырая)  :", repr(r["bta_date_raw"]))
print("  Профилактика      :", repr(r.get("prophylaxis_text")))
print("  Микробиология     :", repr(r.get("micro_text")))
print("  ИМВП за 5 дней    :", repr(r.get("uti_5d_text")))
````

А теперь прогоним эту же строку через парсеры по одному — и увидим, как текст
превращается в данные:

````python
bta   = P.fix_year_typo(P.parse_date(r["bta_date_raw"]))   # дата + починка опечаток в годе
birth = P.parse_birth_from_name(r["name_cell"])            # дата рождения из текста ФИО
age, age_src = P.resolve_age(r.get("age_raw"), r.get("name_cell"), bta)

print(f"Дата БТА (распарсена)   : {bta}")
print(f"Дата рождения из ФИО    : {birth}")
print(f"Возраст / откуда взят   : {age} / {age_src}")
print(f"Тип дренирования        : {P.parse_drainage(r['name_cell'])}")

proph = P.classify_prophylaxis(r.get("prophylaxis_text", ""), r.get("micro_text", ""))
print(f"Профилактика            : режим={proph.regimen}, препарат={proph.drug}, "
      f"резистентность={proph.resistant_noted}")

micro = P.parse_microbiology(r.get("micro_text", ""))
print(f"Микробиология           : ОАМ+={micro.oam_positive}, посев+={micro.mi_positive}, "
      f"max_log_cfu={micro.max_log_cfu}, организмы={[o.name for o in micro.organisms]}")

outcome = P.parse_uti_outcome(r.get("uti_5d_text", ""))
print(f"Исход (ИМВП за 5 сут)    : uti_5d={outcome.uti_5d}, censored={outcome.censored}")
````

`features.build_features` делает ровно это для **каждой** строки и плюс считает
производные признаки:
- **интервалы** `oam_to_bta_days`, `mi_to_bta_days` — за сколько дней до БТА
  сделали тест (положительное число = тест был до процедуры);
- **флаги** `significant_bacteriuria` (нагрузка ≥10⁵), `result_dissociation`
  (ОАМ−, посев+), `any_prophylaxis` и др.

Запускаем на всей таблице:

````python
feat = features.build_features(raw)
print(f"После features: {len(feat)} строк, {feat.shape[1]} колонок")
print("Новые аналитические колонки:", list(feat.columns))
feat.head(3)
````

Вот та же «АВДЕЕВ»-строка целиком после парсинга (транспонирована для
читабельности) — сравните с сырым текстом выше:

````python
feat[feat["name_cell"].astype(str).str.startswith("АВДЕЕВ")].iloc[[0]].T
````

## 3. DEDUP — убираем дубликаты процедур

**Единица анализа — процедура**, а не пациент. Один человек мог пройти БТА
несколько раз (это разные процедуры, их оставляем). Но одна и та же процедура
могла попасть в данные дважды (например, из двух файлов) — такие дубли надо
убрать.

Ключ дедупликации объявлен явно: **`фамилия | дата БТА`**. Именно здесь в
реальной работе возникало расхождение «64 vs 66 процедур» — разные ключи дают
разное число. Поэтому ключ зафиксирован, а не выбирается «на глаз».

Сначала **найдём** дубликаты вручную (тем же ключом, что использует `dedup`):

````python
tmp = feat.copy()
tmp["last_name"] = tmp["name_cell"].map(dedup._last_name)   # тот же helper, что внутри dedup
tmp["procedure_key"] = tmp["last_name"] + "|" + tmp["bta_date"].astype(str)

dups = tmp[tmp.duplicated("procedure_key", keep=False)].sort_values("procedure_key")
print(f"Строк до дедупликации : {len(feat)}")
print(f"Задвоенных ключей     : {dups['procedure_key'].nunique()} "
      f"(строк в них: {len(dups)})")
dups[["source_file", "name_cell", "bta_date", "procedure_key"]]
````

> Обратите внимание: например, у БОБРОВА **две** процедуры с *разными* датами —
> это НЕ дубликаты, а два разных визита, оба останутся. Удаляются только строки
> с одинаковым `procedure_key`.

При склейке дубля `dedup` оставляет «более богатую» запись (где больше
заполненных полей). Запускаем:

````python
clean = dedup.deduplicate(feat, bta_date_col="bta_date")
print(f"После дедупликации    : {len(clean)} процедур")
print(f"Удалено дубликатов    : {clean.attrs['n_removed_duplicates']}")
print("Добавлены служебные колонки:", "last_name", "procedure_key")
clean.head(3)
````

## 4. ФИЛЬТР ВКЛЮЧЕНИЯ — критерий по возрасту

Протокол (`config/protocol.yaml`) задаёт критерии включения. Здесь — возраст
≥ 18. Пациентов с **неизвестным** возрастом не выбрасываем (помечаем), а явно
младше порога — исключаем. Параметры берём из конфига, а не «хардкодим».

````python
cfg = pipeline.load_config(ROOT / "config" / "protocol.yaml")
min_age = cfg["study"]["inclusion"]["min_age"]

n_excluded = int(((clean["age"].notna()) & (clean["age"] < min_age)).sum())
clean = clean[(clean["age"].isna()) | (clean["age"] >= min_age)].reset_index(drop=True)

print(f"Критерий включения : возраст ≥ {min_age}")
print(f"Исключено по возрасту: {n_excluded}")
print(f"Осталось процедур   : {len(clean)}")
````

## 5. VALIDATE — контракт данных (Pandera)

Перед статистикой таблица проходит **контракт**: проверку, что все колонки
имеют нужный тип и допустимые значения (даты в разумном диапазоне, `sex` ∈
{м, ж}, `drainage` из списка категорий, `age` 18–100 и т.д.).

Это замена ручному «увидел артефакт глазами → поправил». Если в данные просочится
мусор — пайплайн **упадёт здесь**, а не молча испортит числа.

````python
validated = schema.validate(clean)
print(f"✅ Контракт пройден: таблица валидна, {len(validated)} строк.")
````

Покажем, что контракт действительно **ловит** ошибки. Намеренно испортим одну
ячейку (недопустимая категория дренирования) и попробуем провалидировать —
ожидаем понятное падение:

````python
bad = clean.copy()
bad.loc[bad.index[0], "drainage"] = "банан"     # такой категории нет в контракте

try:
    schema.validate(bad)
    print("(сюда дойти не должны)")
except Exception as e:
    print("❌ Контракт поймал артефакт ещё до статистики:")
    print("   тип ошибки:", type(e).__name__)
    for line in str(e).splitlines()[:5]:
        print("  ", line)
````

## 6. STATS — считаем все числа

На валидной таблице запускаем статистический слой. Каждая функция — отдельная
конечная точка анализа; все они складываются в один объект `Results`.

````python
primary = stats.primary_endpoint(clean)
mc      = stats.oam_vs_mi_mcnemar(clean)
timing  = stats.oam_vs_mi_timing(clean)
fisher  = {f: stats.fisher_factor(clean, f)
           for f in ["any_prophylaxis", "resistant_noted", "significant_bacteriuria"]}

print("Первичная точка : ИМВП", primary["uti_events"], "из", primary["n_evaluable"],
      f"({primary['uti_rate_wilson']['point']:.1%})")
print("McNemar         : ОАМ−/посев+ =", mc["oam_neg_mi_pos"],
      "| p =", round(mc["p_value"], 6))
print("Сроки (медианы) : ОАМ", timing["oam_median"], "сут | посев", timing["mi_median"], "сут")
print("Факторы (Fisher):", {k: round(v["p_value"], 3) for k, v in fisher.items()})
````

## 7. Проверка: ручной прогон == `pipeline.run(...)`

Мы прошли конвейер вручную. Теперь вызовем «боевую» функцию одной строкой и
убедимся, что результат тот же. Именно так анализ, отчёт и слайды берут числа
из **одного** источника — расхождений быть не может.

````python
res = pipeline.run(RAW_DIR, ROOT / "config" / "protocol.yaml")
print("provenance (происхождение чисел):")
for k, v in res.provenance.items():
    print(f"   {k:24s}: {v}")

assert len(res.analytical) == len(clean), "число процедур должно совпасть!"
print(f"\n✅ Ручной прогон и pipeline.run() согласованы: {len(res.analytical)} процедур.")
````

И наглядная **воронка** — как число строк уменьшалось по ходу пайплайна:

````python
prov = res.provenance
stages = ["Сырые строки", "После дедупликации", "Аналитических процедур"]
vals = [prov["n_raw_rows"], prov["n_after_dedup"], prov["n_analytical"]]

fig, ax = plt.subplots(figsize=(7.5, 3.2))
bars = ax.barh(stages, vals, color=["#C44E52", "#DD8452", "#4C72B0"])
for y, v in enumerate(vals):
    ax.text(v + 0.4, y, str(v), va="center", fontweight="bold")
ax.invert_yaxis()
ax.set_xlim(0, max(vals) + 6)
ax.set_xlabel("Число строк / процедур")
ax.set_title("Воронка пайплайна: от сырых данных к аналитике")
plt.tight_layout(); plt.show()
````

## Итого — что мы увидели

| Шаг | Слой | Было → стало | Что делает |
|-----|------|--------------|------------|
| 1 | `ingest`   | 2 файла → 1 таблица | читает Excel, переименовывает, склеивает |
| 2 | `features` | текст → колонки | парсит даты/возраст/микробы, считает интервалы и исходы |
| 3 | `dedup`    | 68 → 64 | убирает дубли процедур по ключу `фамилия|дата` |
| 4 | фильтр     | критерий 18+ | исключает не подходящих по возрасту |
| 5 | `schema`   | проверка | контракт ловит артефакты до статистики |
| 6 | `stats`    | таблица → числа | доли, McNemar, Wilcoxon, Fisher, динамика |

Каждый слой — чистая, тестируемая функция. Поэтому пайплайн воспроизводим, а
любое изменение протокола (`config/protocol.yaml`) перепроливает весь анализ
без правок кода.
